```python
def make_wc(keyword_list):
    wc = wordcloud.WordCloud(relative_scaling=.5,prefer_horizontal=.75,max_words=200,min_font_size=8)
    wc1 = wc.fit_words(keyword_list)
    actualwc = wc1.to_array()
    plt.title("Wordcloud Sample")
    plt.imshow(actualwc)
    plt.axis("off")
    plt.show()
```

```python
fig = make_wc(keyword_sample)
```
